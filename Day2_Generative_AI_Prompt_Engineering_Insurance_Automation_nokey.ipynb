# 2. LLM Fundamentals for Automation Teams

A Large Language Model, or LLM, is an AI model that can understand and generate text.

For automation teams, LLMs are useful for:

- Summarizing claim notes
- Extracting information from unstructured text
- Drafting customer emails
- Explaining policy terms
- Classifying requests
- Creating structured JSON for downstream automation
- Supporting UiPath workflows with intelligence

## Traditional Automation vs AI Automation

| Traditional UiPath Automation | AI-Enabled Automation |
|---|---|
| Works best with structured data | Can handle unstructured text |
| Rule-based | Language and context-aware |
| Predictable but rigid | Flexible but needs guardrails |
| Reads fields and tables | Understands meaning |
| Needs exact conditions | Can reason from messy notes |

**Trainer Humour:** Traditional bot says, “Give me exact field names.” AI bot says, “Give me messy notes also.” Enterprise trainer says, “Good. Now show the audit trail.”

# 4. Mock LLM Function for Classroom Practice

In real projects, you may use OpenAI, Azure OpenAI, AWS Bedrock, Google Gemini, Anthropic Claude or an internal enterprise LLM gateway.

For beginner-friendly practice, we first use a mock function. This allows participants to focus on prompt design without API setup issues.

````python
def mock_llm_response(task, prompt):
    """Mock response generator for classroom practice. It does not call any external API."""
    task = task.lower()

    if task == "claim_summary":
        return """Claim Summary:
Rajesh Sharma submitted a health insurance claim for dengue treatment at City Care Hospital.
The claim amount is INR 85,000 under policy POL1001. Submitted documents include hospital bill, discharge summary and doctor prescription. Current status: pending document verification."""

    if task == "policy_explanation":
        return """Policy Coverage Explanation:
The policy covers hospitalization expenses up to INR 5,00,000. Room rent is covered up to INR 5,000 per day. It also covers pre-hospitalization expenses for 30 days and post-hospitalization expenses for 60 days. Cosmetic and experimental treatments are excluded."""

    if task == "customer_email":
        return """Dear Rajesh Sharma,

Thank you for contacting us. We have received your health insurance claim request. Your claim is currently under document verification. Please ensure that your hospital bill, discharge summary and doctor prescription are submitted.

We will update you once the verification is completed.

Regards,
Claims Support Team"""

    if task == "risk_indicators":
        return """Risk Indicators:
1. Policy started only 5 days before the accident.
2. Police report is missing.
3. Photos are unclear.
4. Repair estimate appears high for minor damage.
5. Accident happened at midnight.

Risk Level: High
Recommended Next Action: Request police report, clearer photos and garage estimate verification."""

    if task == "json_output":
        return {
            "claim_id": "CLM9001",
            "policy_number": "POL1001",
            "customer_name": "Rajesh Sharma",
            "claim_type": "Health",
            "claim_amount": 85000,
            "summary": "Health insurance claim for dengue treatment at City Care Hospital.",
            "status": "Pending Document Verification",
            "next_action": "Verify submitted documents"
        }

    return "Please make the task more specific."

print(mock_llm_response("claim_summary", "Create claim summary"))
````

# 6. Prompt Engineering Pattern 1: Clear Task Prompt

A clear prompt tells the LLM:

1. What role to play
2. What task to perform
3. What input to use
4. What output format is expected

## Weak Prompt

```text
Summarize this.
```

## Better Prompt

```text
You are an insurance claims assistant.
Create a 5-line claim summary from the claim notes.
Mention customer name, claim type, claim amount, documents and current status.
```

````python
prompt = f"""
You are an insurance claims assistant.

Task:
Create a concise claim summary from the raw claim notes.

Include:
- Customer name
- Claim type
- Claim amount
- Hospital name
- Submitted documents
- Current status

Raw Claim Notes:{raw_claim_notes}
"""

response = mock_llm_response("claim_summary", prompt)
print(response)
````

# 7. Prompt Engineering Pattern 2: Role + Context + Output

This pattern is useful in enterprise use cases.

## Template

```text
Role:
You are a [specific role].

Context:
You are working for [business process].

Task:
Perform [specific task].

Output:
Return [format].
```

This reduces vague answers and improves repeatability.

````python
prompt = f"""
Role:
You are a senior insurance policy servicing assistant.

Context:
A customer wants to understand health policy coverage in simple language.

Task:
Explain the policy coverage clearly.

Output:
Use simple bullet points. Avoid legal jargon.

Policy Coverage:
{policy_coverage_text}
"""

print(mock_llm_response("policy_explanation", prompt))
````

# 9. System, User and Developer-Style Instructions

In many LLM applications, instructions can be organized into layers.

## System Instruction

High-level behavior.

```text
You are a professional insurance automation assistant.
Follow compliance-friendly language.
Do not approve or reject claims without evidence.
```

## Developer-Style Instruction

Application-specific rules.

```text
Always return claim outputs in JSON format.
Always include missing_documents field.
```

## User Instruction

Task-specific request.

```text
Summarize this claim note.
```

For UiPath developers, this is similar to separating process rules, business logic and transaction-specific data.

````python
system_instruction = """
You are a professional insurance automation assistant.
Use clear, neutral and compliance-friendly language.
Do not make final claim approval decisions without policy evidence.
"""

developer_instruction = """
Always structure the answer for downstream automation.
If documents are missing, mention them clearly.
Avoid emotional or promotional language.
"""

user_instruction = f"""
Create a claim summary from the following notes:

{raw_claim_notes}
"""

combined_prompt = f"""
SYSTEM INSTRUCTION:
{system_instruction}

DEVELOPER INSTRUCTION:
{developer_instruction}

USER REQUEST:
{user_instruction}
"""

print(mock_llm_response("claim_summary", combined_prompt))
````

# 12. Structured Output and JSON Response Design

UiPath works best with structured data.

Instead of asking the LLM to return paragraphs, we can ask for JSON.

## Why JSON?

- Easy for UiPath to parse
- Easy for APIs to consume
- Easy to store in databases
- Easy to validate
- Useful for audit trails

## Important Prompt Rule

```text
Return valid JSON only. Do not add explanation outside JSON.
```

````python
json_prompt = f"""
You are an insurance automation assistant.

Extract information from the claim notes and return JSON only.

Required JSON fields:
- claim_id
- policy_number
- customer_name
- claim_type
- claim_amount
- summary
- status
- next_action

Raw Claim Notes:
{raw_claim_notes}
"""

json_response = mock_llm_response("json_output", json_prompt)
print(json.dumps(json_response, indent=4))
````

# 14. Validating JSON Before UiPath Consumption

LLM output should be validated before passing it to UiPath or any downstream system.

Validation checks:

- Is it valid JSON?
- Are required fields present?
- Are data types correct?
- Is claim amount numeric?
- Is recommended action available?

````python
required_fields = [
    "claim_id", "policy_number", "customer_name", "claim_type", "claim_amount",
    "document_list", "risk_indicators", "missing_documents", "recommended_action"
]

def validate_insurance_json(output):
    errors = []

    for field in required_fields:
        if field not in output:
            errors.append(f"Missing field: {field}")

    if "claim_amount" in output and not isinstance(output["claim_amount"], (int, float)):
        errors.append("claim_amount must be numeric")

    if "document_list" in output and not isinstance(output["document_list"], list):
        errors.append("document_list must be a list")

    if "recommended_action" in output and not output["recommended_action"]:
        errors.append("recommended_action must not be empty")

    return errors

errors = validate_insurance_json(uipath_json_output)

if not errors:
    print("Output passed validation and is ready for UiPath consumption")
else:
    print("Validation errors:")
    for error in errors:
        print("-", error)
````

## Mini Assignment 5: Add More Validation Rules

Enhance the validator to check:

- risk_indicators must be a list
- missing_documents must be a list
- claim_type should be Health, Vehicle, Life, Travel or Property

````python
def advanced_validate_insurance_json(output):
    errors = validate_insurance_json(output)
    allowed_claim_types = ["Health", "Vehicle", "Life", "Travel", "Property"]

    if "risk_indicators" in output and not isinstance(output["risk_indicators"], list):
        errors.append("risk_indicators must be a list")

    if "missing_documents" in output and not isinstance(output["missing_documents"], list):
        errors.append("missing_documents must be a list")

    if "claim_type" in output and output["claim_type"] not in allowed_claim_types:
        errors.append("claim_type is not valid")

    return errors

advanced_errors = advanced_validate_insurance_json(uipath_json_output)
print("Advanced validation errors:", advanced_errors)
````
