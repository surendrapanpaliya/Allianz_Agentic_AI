## Lab 1: Create Reusable Claim Validation Functions

### Task

Create functions for:

1. Policy status validation
2. Premium status validation
3. Claim amount validation
4. Final claim decision

### Expected Output

`Approved for Processing` or `Rejected: Reason`

````python
def validate_claim(policy_status, premium_status, claim_amount, coverage_amount):
    if policy_status != "Active":
        return "Rejected: Policy is not active"

    if premium_status != "Paid":
        return "Rejected: Premium is pending"

    if claim_amount > coverage_amount:
        return "Rejected: Claim amount exceeds coverage amount"

    return "Approved for Processing"


decision = validate_claim(
    policy_status="Active",
    premium_status="Paid",
    claim_amount=85000,
    coverage_amount=500000
)

print(decision)
````

# 8. Exception Handling for Automation Reliability

Automation can fail due to missing files, invalid file paths, missing columns, invalid JSON, API failures and incorrect data types.

Exception handling helps us build reliable automations.

````python
try:
    with open("insurance_lab_data/claim_data.json", "r") as file:
        claim_data = json.load(file)

    print("Claim ID:", claim_data["claim_id"])
    print("Claim Amount:", claim_data["claim_amount"])

except FileNotFoundError:
    print("Error: Claim file not found")

except KeyError as e:
    print("Error: Missing field", e)

except json.JSONDecodeError:
    print("Error: Invalid JSON format")

except Exception as e:
    print("Unexpected error:", e)
````

# 10. Generate Clean Claim Processing Output File

In real automation, the output may be passed to UiPath queue, Excel report, email, API, database, audit system or claim management system.

````python
claim = {
    "claim_id": "CLM9001",
    "policy_number": "POL1001",
    "customer_name": "Rajesh Sharma",
    "claim_amount": 85000
}

policy_details = mock_policy_verification_api(claim["policy_number"])

if policy_details:
    if policy_details["policy_status"] == "Active" and policy_details["premium_status"] == "Paid":
        if claim["claim_amount"] <= policy_details["coverage_amount"]:
            validation_status = "Passed"
            decision = "Approved for Processing"
            rejection_reason = ""
        else:
            validation_status = "Failed"
            decision = "Rejected"
            rejection_reason = "Claim amount exceeds coverage"
    else:
        validation_status = "Failed"
        decision = "Rejected"
        rejection_reason = "Policy inactive or premium pending"
else:
    validation_status = "Failed"
    decision = "Rejected"
    rejection_reason = "Policy not found"

output = {
    "claim_id": claim["claim_id"],
    "policy_number": claim["policy_number"],
    "customer_name": claim["customer_name"],
    "claim_amount": claim["claim_amount"],
    "validation_status": validation_status,
    "decision": decision,
    "rejection_reason": rejection_reason
}

output
````

# 11. End-to-End Lab: Insurance Document Processor

## Business Scenario

An insurance company receives claim requests from customers. The automation system must read claim data, verify policy details, check premium status, validate claim amount and generate a clean processing output file.

````python
def read_policy_master(file_path):
    return pd.read_excel(file_path)


def read_claim_data(file_path):
    with open(file_path, "r") as file:
        return json.load(file)


def find_policy(policy_df, policy_number):
    result = policy_df[policy_df["policy_number"] == policy_number]
    if result.empty:
        return None
    return result.iloc[0].to_dict()


def process_claim(policy, claim):
    if policy is None:
        return {"validation_status": "Failed", "decision": "Rejected", "rejection_reason": "Policy not found"}
    if policy["policy_status"] != "Active":
        return {"validation_status": "Failed", "decision": "Rejected", "rejection_reason": "Policy inactive"}
    if policy["premium_status"] != "Paid":
        return {"validation_status": "Failed", "decision": "Rejected", "rejection_reason": "Premium pending"}
    if claim["claim_amount"] > policy["coverage_amount"]:
        return {"validation_status": "Failed", "decision": "Rejected", "rejection_reason": "Claim amount exceeds coverage"}
    return {"validation_status": "Passed", "decision": "Approved for Processing", "rejection_reason": ""}


def save_output(output, file_path):
    with open(file_path, "w") as file:
        json.dump(output, file, indent=4)
````

````python
policy_master = read_policy_master("insurance_lab_data/policy_master.xlsx")
claim_data = read_claim_data("insurance_lab_data/claim_data.json")

policy = find_policy(policy_master, claim_data["policy_number"])
decision_result = process_claim(policy, claim_data)

final_output = {
    "claim_id": claim_data["claim_id"],
    "policy_number": claim_data["policy_number"],
    "customer_name": claim_data["customer_name"],
    "claim_amount": claim_data["claim_amount"],
    "validation_status": decision_result["validation_status"],
    "decision": decision_result["decision"],
    "rejection_reason": decision_result["rejection_reason"]
}

save_output(final_output, "insurance_lab_data/final_claim_output.json")
final_output
````
